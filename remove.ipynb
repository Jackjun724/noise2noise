```python
from PIL import Image
import random
```

```python
image = Image.open("test.jpg")
image
```

```python
watermark = Image.open('./mask-without-border-clip.png')  # 水印路径
watermark.mode
```

```python
if watermark.mode != 'RGBA':
    l_channel, a_channel = watermark.split()
    watermark = Image.merge("RGBA", (l_channel, l_channel, l_channel, a_channel))
    print('add alpha tunnel')
watermark
```

```python
paste_mask = watermark.split()[3].point(lambda i: i * (35 + random.randint(-5, 5)) / 100.)
image.paste(watermark, (0, 0), mask=paste_mask)
image
```
